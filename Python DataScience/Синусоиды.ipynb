```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import math as mt
import random as rnd
import statistics as sts
from IPython.display import display
```

```python
def draw_graphic(xrow, yrow):
    plt.figure(figsize=(30, 20))
    plt.rc('xtick', labelsize=22)    
    plt.rc('ytick', labelsize=22)
    plt.grid(color='k')
    plt.minorticks_on()
    plt.grid(which='minor', 
            color = 'k', 
            linestyle = ':')
    plt.plot(xrow, yrow, linewidth=3)
    plt.xlabel('Time', fontsize=24)
    plt.ylabel('Amplitude', fontsize=24)
    plt.title('График', fontsize=28)
    plt.show()
```

```python
def draw(xrow, yrow):
 
    fig, ax = plt.subplots() 

    fig.set_figwidth(14) 
    fig.set_figheight(8) 

    ax.plot(xrow,yrow, linestyle='-') 

    ax.set_xlabel('Время', fontsize=15) 
    ax.set_ylabel('Амплитуда', fontsize=15) 
    ax.set_title(f'Идеальный график', fontsize=15) 

    #  Устанавливаем интервал основных делений: 
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1)) 
    ax.yaxis.set_major_locator(ticker.MultipleLocator(10)) 

    #  Устанавливаем форматирование чисел основных делений: 
    ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f')) 
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f')) 

    #  Устанавливаем форматирование делений: 
    ax.xaxis.set_tick_params(which = 'major', labelsize = 15, labelrotation = 45) 
    ax.yaxis.set_tick_params(which = 'major', labelsize = 15) 

    #item.set_ylim(-0.05, 1.05) 
    #ax.set_ylim(min(df[col]) - step, max(df[col]) + step) 

    # Вывод и настройка сетки 
    ax.minorticks_on() 
    ax.grid(which='major', color = 'k', alpha = 0.6) 
    ax.grid(which='minor', color = 'gray', linestyle = ':') 

    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4) 
    plt.show()
```

```python
def break_row(xrow, yrow):
    break_x = xrow.copy()
    break_y = yrow.copy()
    time_fit = rnd.uniform(0.1,0.2)
    k=1
    rnd_list = [1,2]
    start = 24
    border = rnd.randint(22,27)
    middle_value = sts.mean(xrow[:22])
    noise_num = rnd.choice(rnd_list)
    for i in range(4):
        for j in range(noise_num):
            multipiler = rnd.uniform(-1.2,1.2)
            break_y.insert(start+j, round(middle_value*multipiler, 2))
            break_x.insert(start+j, round(xrow[start-1]+time_fit*k, 2))
            k+=1
        start-=6
    break_x = break_x[:32]
    break_y = break_y[:32]
    return break_x, break_y
```

```python
def build_row(start_space=0, left_border=40 , right_border=45, lb_between_group = 2.5, rb_between_group = 4.3, 
              max_amplitude=60, min_amplitude=30, par1 = 1.2, par2 = 7.3):
    #
    end_space = round(rnd.uniform(left_border, right_border),1)
    x = np.linspace(start_space, end_space, 2000)
    dlt_t = round(rnd.uniform(0.1,2), 1)
    yarr = []
    xarr = []
    xrow = []
    yrow = []
    
    last_point = 0
    
    distance_btw_group = round(rnd.uniform(lb_between_group, rb_between_group), 1)
    beta = np.pi/dlt_t
    alpha = 6*np.pi/dlt_t
    
    fi1 = round(rnd.uniform(24, 27), 1)
    si1 = fi1+distance_btw_group
    ti1 = si1+distance_btw_group
    fit1 = ti1+distance_btw_group
    
    first_interval =[fi1, fi1 + dlt_t]
    second_interval =[si1, si1 + dlt_t]
    third_interval =[ti1, ti1 + dlt_t]
    fourth_interval =[fit1 , fit1 + dlt_t]
    
    if(end_space<fit1 + dlt_t ):
         x = np.linspace(start_space, fit1 + dlt_t+1, 2000)
    #par1 = fi1 + dlt_t/2
    #par2 = fit1 + dlt_t/2
    koef = (mt.log(max_amplitude/min_amplitude))/(-par1+par2)
    const_amp = max_amplitude/np.exp(-koef*par1)
    for i in range(len(x)):

        if( first_interval[0] < x[i] <  first_interval[1]):
            signal = const_amp*np.exp(-koef*(x[i]-25))*np.sin((x[i]-first_interval[0])*beta)*np.sin((x[i]-first_interval[0])*alpha)
            
            yarr.append(signal)
            xarr.append(x[i])

        elif(second_interval[0] < x[i] <  second_interval[1]):
            signal = const_amp*np.exp(-koef*(x[i]-25))*np.sin((x[i]-second_interval[0])*beta)*np.sin((x[i]-second_interval[0])*alpha)
                                                               
            yarr.append(signal)
            xarr.append(x[i])

        elif(third_interval[0] < x[i] <  third_interval[1]):
            signal = const_amp*np.exp(-koef*(x[i]-25))*np.sin((x[i]-third_interval[0])*beta)*np.sin((x[i]-third_interval[0])*alpha)
                                                               
            yarr.append(signal)
            xarr.append(x[i])

        elif(fourth_interval[0] < x[i] <  fourth_interval[1]):
            signal = const_amp*np.exp(-koef*(x[i]-25))*np.sin((x[i]-fourth_interval[0])*beta)*np.sin((x[i]-fourth_interval[0])*alpha)
                                                               
            yarr.append(signal)
            xarr.append(x[i])
            last_point = i
            
    for i in range (1, len(yarr)-1):
    
        if(yarr[i+1]<yarr[i] and yarr[i-1]<yarr[i]):
            yrow.append(round(yarr[i], 2))
            xrow.append(round(xarr[i], 2))
            
            
            
        if(yarr[i+1]>yarr[i] and yarr[i-1]>yarr[i]):
            yrow.append(round(yarr[i], 2))
            xrow.append(round(xarr[i], 2))
            
            
            
    for i in range (8):
        xrow.append(round(1+x[last_point+i],2))
        yrow.append(0)
        
    return xrow, yrow
```

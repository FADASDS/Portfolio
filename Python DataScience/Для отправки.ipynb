```python
import pandas as pd
import numpy as np
from numpy import sqrt
import matplotlib.pyplot as plt
import statsmodels.api as sm
import sys
sys.path.append('custom_modules')
import Data_processing as dp
import Data_visualization as dv
```

```python
def draw_4_graphs1(df,df1):
    plt.figure(figsize=(22, 20))
    i = 1
    columns = list(df.columns[1:])
    
    for col in columns:
        plt.subplot(4, 1, i)
        plt.plot(df['Time Moment'] , df[col], '-',label='Данные до обработки')
        plt.plot(df1['Time Moment'] , df1[col], '-',label='Данные после обработки')
        plt.xlabel('Time Moment')
        plt.ylabel(col)
        plt.title(col)
        i+=1
        plt.legend()
    plt.show()
```

```python
df=dp.get_df('E:/4 семестр/ГПО/РПАС/Данные СТМ Восточный Купол (от 29.01.17)/Куст 202/Скважина 2022.xlsx')
df=dp.normalize_data(df)
Prepared_DF=dp.delete_NANs(df)
Prepared_DF=dp.delete_zeros(Prepared_DF)
Prepared_DF=dp.delete_outliers(Prepared_DF, soft_mode = True)
draw_4_graphs1(df,Prepared_DF)
```
